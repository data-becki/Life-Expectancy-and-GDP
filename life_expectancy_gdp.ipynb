```python
year = df['Year'].unique()
year
```

```python
chile_le = df['Life expectancy at birth (years)'][df['Country'] == 'Chile']
# chile_le
china_le = df['Life expectancy at birth (years)'][df['Country'] == 'China']
ger_le = df['Life expectancy at birth (years)'][df['Country'] == 'Germany']
mex_le = df['Life expectancy at birth (years)'][df['Country'] == 'Mexico']
us_le = df['Life expectancy at birth (years)'][df['Country'] == 'United States of America']
zim_le = df['Life expectancy at birth (years)'][df['Country'] == 'Zimbabwe']
```

```python
# Define data values
# x = [7, 14, 21, 28, 35, 42, 49] # Year
# y = [5, 12, 19, 21, 31, 27, 35] # Chile's life expectancy
# z = [3, 5, 11, 20, 15, 29, 31] # next country's life expectancy

# Plot a simple line chart
plt.plot(year, chile_le)
plt.plot(year, china_le)
plt.plot(year, ger_le)
plt.plot(year, mex_le)
plt.plot(year, us_le)
plt.plot(year, zim_le)
plt.legend(df['Country'].unique())
plt.title('Life Expectancy over time')
plt.show()
```

*The line chart doesn't show enough detail in the top countries. Try subplots.*

```python
# First Subplot
plt.subplot(2, 3, 1)
plt.plot(year, china_le)
plt.title('China')
 
# Second Subplot
plt.subplot(2, 3, 2)
plt.plot(year, chile_le)
plt.title('Chile')

# Third Subplot
plt.subplot(2, 3, 3)
plt.plot(year, ger_le)
plt.title('Germany')

# Fourth Subplot
plt.subplot(2, 3, 4)
plt.plot(year, mex_le)
plt.title('Mexico')
 
# Second Subplot
plt.subplot(2, 3, 5)
plt.plot(year, us_le)
plt.title('USA')

# Third Subplot
plt.subplot(2, 3, 6)
plt.plot(year, zim_le)
plt.title('Zimbabwe')
 
# Display both subplots
# plt.figure(figsize=(100, 100))
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
```

### How has GDP changed over time?

```python
chile_gdp = df['GDP'][df['Country'] == 'Chile']
# chile_le
china_gdp = df['GDP'][df['Country'] == 'China']
ger_gdp = df['GDP'][df['Country'] == 'Germany']
mex_gdp = df['GDP'][df['Country'] == 'Mexico']
us_gdp = df['GDP'][df['Country'] == 'United States of America']
zim_gdp = df['GDP'][df['Country'] == 'Zimbabwe']
```

```python
plt.plot(year, chile_gdp)
plt.plot(year, china_gdp)
plt.plot(year, ger_gdp)
plt.plot(year, mex_gdp)
plt.plot(year, us_gdp)
plt.plot(year, zim_gdp)
plt.legend(df['Country'].unique())
plt.title('GDP over time')
plt.show()
```

*We have the same issue with the previous chart. Split the information into subplots.*

*Life expectancy has increased over time, and the GDP has increased over time, in the six countries studied.* 

### What is the average life expectancy in these nations?

```python
df_means = df.drop('Year', axis=1).groupby('Country').mean().reset_index()
df_means
```

```python
ax1=sns.barplot(x = 'Country', y = 'Life expectancy at birth (years)', data = df_means)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
ax2=sns.barplot(x = 'Country', y = 'GDP', data = df_means)
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
ax1=sns.violinplot(x = 'Country', y = 'Life expectancy at birth (years)', data = df)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
ax2=sns.violinplot(x = 'Country', y = 'GDP', data = df)
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```
